Correlation analysis observations from this heatmap:

1. There is a strong positive relationship (0.81) between population and total bedrooms as it is true, when the population increases, the numbers of bedrooms increase, and vice versa.

2. There is a medium-strong positive relationship (0.69) between Median House Value and Median income, as the income level would increase by an increase in the median house value.

3. There is a strong positive relationship (0.85) between housholds and the popluation values as it is true, with an increase in population, number of houses will increase, but thias only is accurate in one way of the correlation.

4. There is a weak negative relationship (-0.1) between total bedrooms and the median income as it is true, those who have less income will have lesser number of bedrooms based on low cost affordability.

5. There is a weak negative relationship (-0.27) between housing median age and the total rooms as it is true, in previous years due to the size of the family being smaller than it is today, less number of rooms were required to live in.
